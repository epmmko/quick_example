```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x_file = pd.read_csv("x_val.csv", header=None)
h_file = pd.read_csv("h_val.csv", header=None)
y_file = pd.read_csv("GT_LL.csv", header=None)
```

```python
x_val = x_file.values[0,:]
h_val = h_file.values[:,0]
xx = x_file.values
hh = h_file.values
yy = y_file.values
levels = [-10000,-2000,-1000,-500,-200,-100,-40,-10,0,3,4,5,7,
          10,20,40,100,200,500,1000,2000,10000,20000]
```

```python
font = {'size': 20}
mpl.rc('font', **font)

plt.figure(figsize=(16,9))
ax = plt.subplot(111)
ax.set_xscale('log')
contour = plt.contour(xx,hh,yy, levels=levels , linewidths=2, cmap=plt.cm.jet,
                       norm=mpl.colors.SymLogNorm(linthresh = 0.1,vmin=yy.min(), vmax=yy.max()))
plt.clabel(contour, inline = True, fmt='%3.0f', fontsize=18, colors='black')
plt.xlim(0.0009,110)
plt.ylim(-0.01,1.01)
ax.set_yticks(np.linspace(0,1,21),minor=True)
plt.xlabel('X')
plt.ylabel('h/D')
plt.grid(b=True,which='major',color='k',linestyle='-',axis='both')
plt.grid(b=True,which='minor',color='b',linestyle='--',axis='both',alpha = 0.5)
plt.title("Laminar Liquid, Turbulent Gas")
plt.savefig("GT_LL.png")
plt.show()
```
